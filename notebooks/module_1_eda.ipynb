# VOLTEX — Module 1 EDA

Checks historical surge prevalence and confirms that known market-stress dates appear as volume-z-score spikes.

```python
from pathlib import Path
import sys
import matplotlib.pyplot as plt
import pandas as pd

ROOT = Path.cwd().resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))
from src.data.features import build_historical_features

csv_path = next((ROOT / 'data').glob('*.csv'))
features = build_historical_features(csv_path)
features.head()
```

```python
prevalence = features['surge_label'].mean()
print(f'Rows: {len(features):,}')
print(f'Surge prevalence: {prevalence:.2%}')

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(features['date'], features['volume_zscore_20d'], lw=0.9, color='#00B4D8')
ax.axhline(2, color='#E74C3C', ls='--', lw=1, label='2σ')
for label, day in {'Black Monday': '2015-08-24', 'Brexit': '2016-06-24', 'VIX shock': '2018-02-05'}.items():
    ax.axvline(pd.Timestamp(day), color='#F0A500', alpha=.7, lw=1)
    ax.text(pd.Timestamp(day), ax.get_ylim()[1] * .9, label, rotation=90, va='top', fontsize=8)
ax.set(title='Lagged 20-day volume z-score', ylabel='z-score', xlabel='Prediction date')
ax.legend(); plt.show()
```
